```python
import os
import json
import skimage

json_path = "/content/gdrive/MyDrive/ECM_data/trainval_final.json"
images_path = "/content/gdrive/MyDrive/ECM_data/images"
```

```python
images_list = os.listdir(images_path)
```

```python
with open(json_path) as json_file:
    data = json.load(json_file)
```

```python
!mkdir output_images
!pwd
```

```python
for image in data["images"]:
  filename = image["file_name"]
  pic = skimage.io.imread(os.path.join(images_path, filename))
  # print(filename)
  skimage.io.imsave(os.path.join("./output_images", filename), pic)
```
